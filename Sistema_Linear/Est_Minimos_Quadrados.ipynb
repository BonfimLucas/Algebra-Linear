# **ESTIMATIVA POR MINIMOS QUADRADOS**

```python
# descobrindo a transposta da matriz a
at = np.array(a).T
at
```

```python
a
```

```python
# Multiplicação de transposta pela matriz a
ata = np.dot(at,a)
ata
```

```python
# calcular a inversa da matriz anterior
inv = np.linalg.inv(ata)
inv
```

```python
atb = np.dot(at,b)
atb
```

```python
x = np.dot(inv,atb)
x
```

```python
#clean code
x = np.dot(np.dot(np.linalg.inv(np.dot(at,a)),at),b)
x
```

**nota = 0.1839.inicio_estudo + 0.0414.tempo_estudo_dia**

```python
resultado = nota
resultado
```

```python
resultado['previsao'] = resultado['inicio_estudo']*x[0]+resultado['tempo_estudo_dia']*x[1]
resultado
```

```python
# Erro medio absoluto (MAE)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(resultado.nota, resultado.previsao)
```

```python
resultado['erro_abs'] = abs (resultado.nota - resultado.previsao)
resultado
```
